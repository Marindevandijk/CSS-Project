```python
import matplotlib.pyplot as plt
```

```python
m_values = []
steps_values = []

with open("seed0steps.txt", "r") as f:
    for line in f:
        parts = line.strip().split(",")
        
        m = float(parts[0].split(":")[1])
        steps = int(parts[1].split(":")[1])
        
        m_values.append(m)
        steps_values.append(steps)
```

```python
plt.figure()
plt.plot(m_values, steps_values, marker='o')
plt.xlabel("m")
plt.ylabel("steps_til_consensus")
plt.yscale("log")
plt.title("Steps to Consensus vs m")
plt.grid(True)
plt.savefig("stepstoconcensus.png", dpi=300)
plt.show()
```
